## 1.

Para modelar la relaci√≥n de enfriamiento en t√©rminos de una sucesi√≥n recursiva, se puede usar la Ley de Enfriamiento de Newton, la cual establece que la velocidad de cambio de la temperatura de un objeto es proporcional a la diferencia entre su temperatura y la temperatura del ambiente. La ecuaci√≥n recursiva que describe la evoluci√≥n de la temperatura podr√≠a escribirse como:

$T_{n+1} = T_n - k(T_n - T_{ambiente})$

Representando cada variable, al aterrizarse al contexto:
* $T_n$ representa la temperatura del agua en un instante $n$
* $T_ambiente$ es la temperaturo del entorno, es decir, una constante con valor de $22 ¬∞C$
* $k$ es la constante de enfriamiento, la cual es dependiente de las propiedades del objeto y el ambiente, relacionandose con la conductividad terminca y la naturaleza del material utilizado.
* $T_{n+1}$ es la temperatura en el instante despues de una unidad de tiempo adicional.


Esta ecuaci√≥n expresa que la temperatura en el siguiente instante depende de la temperatura actual, menos una fracci√≥n de la diferencia entre la temperatura actual y la del ambiente. Esto refleja el hecho de que el objeto pierde calor m√°s r√°pidamente cuanto mayor es la diferencia con el ambiente.


## 2.
###2.1

Para encontrar $k$, se utilizan mediciones experimentales de dos puntos y la ley de enfriamiento de newton. Para esto se utiliza:

$T_2= 37.7¬∞C$ (temperatura inicial)

$T_4 = 35.7¬∞C$ (Temperatura en un momento posterior)

$k = \frac{T_n - T_{n+1}}{T_n - T_{\text{ambiente}}}$

La f√≥rmula es una consecuencia directa de la ley de enfriamiento de Newton, que establece que la tasa de cambio de la temperatura de un medida y la de el entorno. En este laboratorio se aplica una forma de la ley de newton discreta con el fin de medir intervalos discretos.  

**Se toman un par de valores de temperatura:**

$( T_2 = 37.7,^\circ C )$
$( T_4 = 35.7,^\circ C )$

**Se aplica la f√≥rmula:**

$k = \frac{37.7 - 35.7}{37.7 - 22} = \frac{2}{15.7} \approx 0.1274$

**Por lo tanto:**

$k \approx 0.1274$

Este valor indica que se pierde aproximadamente 12.7% de la diferencia entre su temperatura y la temperatura ambiente, en cada intervalo de tiempo considerado.

##2.2

Forma num√©rica **sucesi√≥n recursiva**:
Se utiliza la ecuaci√≥n recursiva con el fin de calcular la temperatura en un momento posterior  a partir de la temperatura actual y de la constante $k$

Se sabe que la temperatura en el minuto 12 es:

**Dato inicial:**  $T_{12} = 32.2 ,^\circ C $

Se quiere calcular la temperatura en el minuto 14 por lo que se usa la ecuaci√≥n recursiva y se sustituyen los valores conocidos.

**Para calcula:**  $T_{14} = T_{12} - k (T_{12} - T_{\text{ambiente}})$

Sustituyendo los valores:

$ T_{14} = 32.2 - 0.1274 (32.2 - 22) $

$T_{14} = 32.2 - 0.1274 \times 10.2 $

$T_{14} \approx 32.2 - 1.3 $

$T_{14} \approx 30.9 ,^\circ C $

Repetimos el proceso para encontrar $T_{16}$:

$T_{16} = T_{14} - k (T_{14} - T_{\text{ambiente}})$

**Sustituyendo los valores:**

$T_{16} = 30.9 - 0.1274 (30.9 - 22)$

$T_{16} = 30.9 - 0.1274 \times 8.9$

$T_{16} \approx 30.9 - 1.1$

$T_{16} \approx 29.8 ,^\circ C$

**F√≥rmula general:**
La soluci√≥n "exacta" de la Ley de Enfriamiento de Newton tiene la forma:

$T(t) = T_{\text{ambiente}} + (T_0 - T_{\text{ambiente}}) e^{-kt}$

**Aplicamos los valores:**

 * $T_{\text{ambiente}} = 22 ,^\circ C$

* $T_0 = 43.7 ,^\circ C$

* $k \approx 0.1274$

* $t = 16{min}$

$T(16) = 22 + (43.7 - 22) e^{-0.1274 \times 16}$

$T(16) = 22 + 21.7 e^{-2.0384}$

$T(16) \approx 22 + 21.7 \times 0.1303$

$T(16) \approx 22 + 2.83$

$T(16) \approx 24.83 ,^\circ C$

En base a este ejercicio se puede evidenciar como con el m√©todo recursivo se calcula la temperatura en intervalos de tiempo espec√≠ficos, sin embargo es ideal cuando se tienen datos discretos. Por otro lado, la formula general, proporcina una soluci√≥n continua para cualquier tiempo $t$ por lo que es mejor para datos exactos.

## 2.3
###Comportamiento asint√≥tico de la temperatura:

La Ley de Enfriamiento de Newton describe c√≥mo un objeto pierde calor cuando se encuentra en un entorno con una temperatura ambiente constante. Esta ley predice que, con el tiempo, la temperatura del objeto tiende a igualarse con la del entorno, aunque este proceso nunca ocurre instant√°neamente, sino que sigue un comportamiento asint√≥tico.

A medida que el tiempo $t$ avanza, el t√©rmino exponencial $e^{-kt}$ se aproxima cada vez m√°s a cero, lo que implica que la temperatura del objeto $T(t)$ se acerca a la temperatura ambiente $T_{ambiente}$. Este fen√≥meno de acercamiento progresivo sin llegar a un valor exacto en un tiempo finito se conoce como comportamiento asint√≥tico.

En t√©rminos pr√°cticos, esto significa que, aunque el objeto nunca alcanzar√° exactamente la temperatura ambiente en un periodo de tiempo finito, despu√©s de un tiempo suficientemente largo, la diferencia entre la temperatura del objeto y la del ambiente ser√° tan peque√±a que puede considerarse despreciable. Segun lo cual:

$T(\infty) = T_{\text{ambiente}} = 22 ,^\circ C$

Este comportamiento refleja la naturaleza del enfriamiento: cuanto m√°s cerca est√© la temperatura del objeto de la del ambiente, m√°s lento ser√° el enfriamiento, hasta que la diferencia entre ambas temperaturas sea pr√°cticamente imperceptible.

**Verificaci√≥n num√©rica:**

Simulando num√©ricamente con la ecuaci√≥n recursiva, ver√°s que los valores se estabilizan alrededor de 22¬∞C, lo que confirma que transcurrido mucho tiempo la temperatura se iguala a la temperatura ambiente.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos
tiempo = np.array([0, 2, 4, 6, 8, 10, 12])
temperatura = np.array([43.7, 37.7, 35.7, 34.4, 33.2, 32.9, 32.2])
temp_ambiente = 22

# Funci√≥n de enfriamiento de Newton
def enfriamiento_newton(t, k, T0):
    return temp_ambiente + (T0 - temp_ambiente) * np.exp(-k * t)

# Ajustar los par√°metros
popt, _ = curve_fit(enfriamiento_newton, tiempo, temperatura)
k, T0 = popt

print(f"Constante de enfriamiento k: {k:.4f} min^-1")
print(f"Temperatura inicial T0 (ajustada): {T0:.2f}¬∞C")
print(f"Temperatura inicial real: 43.7¬∞C")

# Calcular temperatura a los 16 minutos
T_16 = enfriamiento_newton(16, k, 43.7)
print(f"Temperatura despu√©s de 16 minutos: {T_16:.2f}¬∞C")

# Calcular temperatura a largo plazo
t_largo_plazo = 1000  # Un tiempo muy grande
T_largo_plazo = enfriamiento_newton(t_largo_plazo, k, 43.7)
print(f"Temperatura a largo plazo: {T_largo_plazo:.2f}¬∞C")

# Generar puntos para la curva de ajuste
t_fit = np.linspace(0, 60, 200)
temp_fit = enfriamiento_newton(t_fit, k, 43.7)

# Graficar
plt.figure(figsize=(12, 6))
plt.plot(tiempo, temperatura, 'bo', label='Datos experimentales')
plt.plot(t_fit, temp_fit, 'r-', label='Modelo de enfriamiento')
plt.axhline(y=temp_ambiente, color='g', linestyle='--', label='Temperatura ambiente')
plt.title('Enfriamiento')
plt.xlabel('Tiempo (minutos)')
plt.ylabel('Temperatura (¬∞C)')
plt.legend()
plt.grid(True)
plt.show()

# Funci√≥n para la sucesi√≥n
def sucesion_temperatura(n):
    return temp_ambiente + (43.7 - temp_ambiente) * (1 - k)**n

# Calcular y mostrar los primeros t√©rminos de la sucesi√≥n
n_terms = 10
for n in range(n_terms):
    print(f"T_{n} = {sucesion_temperatura(n):.2f}¬∞C")



# Calcular el tiempo para acercarse a 22.1¬∞C (muy cerca de la temperatura ambiente)
tiempo_cercano = np.log((22.1 - temp_ambiente) / (43.7 - temp_ambiente)) / -k
print(f"\nTiempo para acercarse a 22.1¬∞C: {tiempo_cercano:.2f} minutos")
```

En la gr√°fica que compartiste, se puede observar la evoluci√≥n de la temperatura de un sistema a lo largo del tiempo, comparando los datos experimentales (puntos azules) con un modelo te√≥rico de enfriamiento basado en la Ley de Enfriamiento de Newton (curva roja). En el eje x se representa el tiempo en minutos, mostrando c√≥mo var√≠a la temperatura del sistema en el eje y, medido en grados Celsius. Los puntos azules corresponden a los datos experimentales que reflejan la disminuci√≥n de la temperatura con el tiempo, mientras que la l√≠nea roja es la curva del modelo te√≥rico ajustado utilizando la constante de enfriamiento  $ k = 0.066 $, que indica un proceso de enfriamiento relativamente r√°pido.

Una constante de enfriamiento $ k = 0.066 $ indica un enfriamiento mucho m√°s r√°pido que en los casos t√≠picos de enfriamiento del agua, que suele estar en el rango de $0.01$ a $0.04$  un valor $ k = 0.066$ se podr√≠a observar en situaciones donde el calor se disipa de manera muy eficiente.

La l√≠nea verde punteada marca la temperatura ambiente $(~22¬∞C)$, que es el valor al cual se aproxima el sistema conforme pasa el tiempo. Este comportamiento es t√≠pico en procesos de enfriamiento donde la temperatura del objeto tiende hacia la temperatura del entorno.

###**Explicaci√≥n sobre alcanzar 10¬∞C o 5¬∞C:**

En el escenario en que se realiz√≥ el experimento, la temperatura ambiente era de 22¬∞C. Dado que las temperaturas de 10¬∞C a 5¬∞C est√°n por debajo de este valor, es imposible que el agua las alcance seg√∫n el modelo planteado. Esto se debe a que, seg√∫n la Ley de Enfriamiento de Newton, la temperatura ambiente representa una as√≠ntota, a la cual la temperatura del agua se acercar√° indefinidamente desde arriba, sin descender en ning√∫n momento por debajo de ella.

Desde una perspectiva termodin√°mica, el hecho de que el agua no pueda alcanzar temperaturas inferiores a la temperatura ambiente se relaciona con el segundo principio de la termodin√°mica, que establece que el calor fluye espont√°neamente desde un cuerpo caliente hacia uno fr√≠o, pero no en la direcci√≥n opuesta. En este caso, dado que el agua es m√°s caliente que el ambiente, transfiere calor al entorno hasta que se alcanza un equilibrio t√©rmico. Este equilibrio se diferencia del modelo matem√°tico de Newton, en el cual la temperatura ambiente es tratada como una as√≠ntota que nunca se alcanza por completo. En la realidad, el agua eventualmente llegar√° a la misma temperatura que el ambiente, momento en el cual el flujo de calor se detiene. Sin un mecanismo activo, como un refrigerador u otro medio de enfriamiento, no es posible extraer calor adicional del agua, por lo que no puede enfriarse por debajo de la temperatura ambiente, ya que ello violar√≠a las leyes termodin√°micas.
