## 1.

Para modelar la relación de enfriamiento en términos de una sucesión recursiva, se puede usar la Ley de Enfriamiento de Newton, la cual establece que la velocidad de cambio de la temperatura de un objeto es proporcional a la diferencia entre su temperatura y la temperatura del ambiente. La ecuación recursiva que describe la evolución de la temperatura podría escribirse como:

$T_{n+1} = T_n - k(T_n - T_{ambiente})$

Representando cada variable, al aterrizarse al contexto:
* $T_n$ representa la temperatura del agua en un instante $n$
* $T_ambiente$ es la temperaturo del entorno, es decir, una constante con valor de $22 °C$
* $k$ es la constante de enfriamiento, la cual es dependiente de las propiedades del objeto y el ambiente, relacionandose con la conductividad terminca y la naturaleza del material utilizado.
* $T_{n+1}$ es la temperatura en el instante despues de una unidad de tiempo adicional.


Esta ecuación expresa que la temperatura en el siguiente instante depende de la temperatura actual, menos una fracción de la diferencia entre la temperatura actual y la del ambiente. Esto refleja el hecho de que el objeto pierde calor más rápidamente cuanto mayor es la diferencia con el ambiente.


## 2.
###2.1

Para encontrar $k$, se utilizan mediciones experimentales de dos puntos y la ley de enfriamiento de newton. Para esto se utiliza:

$T_2= 37.7°C$ (temperatura inicial)

$T_4 = 35.7°C$ (Temperatura en un momento posterior)

$k = \frac{T_n - T_{n+1}}{T_n - T_{\text{ambiente}}}$

La fórmula es una consecuencia directa de la ley de enfriamiento de Newton, que establece que la tasa de cambio de la temperatura de un medida y la de el entorno. En este laboratorio se aplica una forma de la ley de newton discreta con el fin de medir intervalos discretos.  

**Se toman un par de valores de temperatura:**

$( T_2 = 37.7,^\circ C )$
$( T_4 = 35.7,^\circ C )$

**Se aplica la fórmula:**

$k = \frac{37.7 - 35.7}{37.7 - 22} = \frac{2}{15.7} \approx 0.1274$

**Por lo tanto:**

$k \approx 0.1274$

Este valor indica que se pierde aproximadamente 12.7% de la diferencia entre su temperatura y la temperatura ambiente, en cada intervalo de tiempo considerado.

##2.2

Forma numérica **sucesión recursiva**:
Se utiliza la ecuación recursiva con el fin de calcular la temperatura en un momento posterior  a partir de la temperatura actual y de la constante $k$

Se sabe que la temperatura en el minuto 12 es:

**Dato inicial:**  $T_{12} = 32.2 ,^\circ C $

Se quiere calcular la temperatura en el minuto 14 por lo que se usa la ecuación recursiva y se sustituyen los valores conocidos.

**Para calcula:**  $T_{14} = T_{12} - k (T_{12} - T_{\text{ambiente}})$

Sustituyendo los valores:

$ T_{14} = 32.2 - 0.1274 (32.2 - 22) $

$T_{14} = 32.2 - 0.1274 \times 10.2 $

$T_{14} \approx 32.2 - 1.3 $

$T_{14} \approx 30.9 ,^\circ C $

Repetimos el proceso para encontrar $T_{16}$:

$T_{16} = T_{14} - k (T_{14} - T_{\text{ambiente}})$

**Sustituyendo los valores:**

$T_{16} = 30.9 - 0.1274 (30.9 - 22)$

$T_{16} = 30.9 - 0.1274 \times 8.9$

$T_{16} \approx 30.9 - 1.1$

$T_{16} \approx 29.8 ,^\circ C$

**Fórmula general:**
La solución "exacta" de la Ley de Enfriamiento de Newton tiene la forma:

$T(t) = T_{\text{ambiente}} + (T_0 - T_{\text{ambiente}}) e^{-kt}$

**Aplicamos los valores:**

 * $T_{\text{ambiente}} = 22 ,^\circ C$

* $T_0 = 43.7 ,^\circ C$

* $k \approx 0.1274$

* $t = 16{min}$

$T(16) = 22 + (43.7 - 22) e^{-0.1274 \times 16}$

$T(16) = 22 + 21.7 e^{-2.0384}$

$T(16) \approx 22 + 21.7 \times 0.1303$

$T(16) \approx 22 + 2.83$

$T(16) \approx 24.83 ,^\circ C$

En base a este ejercicio se puede evidenciar como con el método recursivo se calcula la temperatura en intervalos de tiempo específicos, sin embargo es ideal cuando se tienen datos discretos. Por otro lado, la formula general, proporcina una solución continua para cualquier tiempo $t$ por lo que es mejor para datos exactos.

## 2.3
###Comportamiento asintótico de la temperatura:

La Ley de Enfriamiento de Newton describe cómo un objeto pierde calor cuando se encuentra en un entorno con una temperatura ambiente constante. Esta ley predice que, con el tiempo, la temperatura del objeto tiende a igualarse con la del entorno, aunque este proceso nunca ocurre instantáneamente, sino que sigue un comportamiento asintótico.

A medida que el tiempo $t$ avanza, el término exponencial $e^{-kt}$ se aproxima cada vez más a cero, lo que implica que la temperatura del objeto $T(t)$ se acerca a la temperatura ambiente $T_{ambiente}$. Este fenómeno de acercamiento progresivo sin llegar a un valor exacto en un tiempo finito se conoce como comportamiento asintótico.

En términos prácticos, esto significa que, aunque el objeto nunca alcanzará exactamente la temperatura ambiente en un periodo de tiempo finito, después de un tiempo suficientemente largo, la diferencia entre la temperatura del objeto y la del ambiente será tan pequeña que puede considerarse despreciable. Segun lo cual:

$T(\infty) = T_{\text{ambiente}} = 22 ,^\circ C$

Este comportamiento refleja la naturaleza del enfriamiento: cuanto más cerca esté la temperatura del objeto de la del ambiente, más lento será el enfriamiento, hasta que la diferencia entre ambas temperaturas sea prácticamente imperceptible.

**Verificación numérica:**

Simulando numéricamente con la ecuación recursiva, verás que los valores se estabilizan alrededor de 22°C, lo que confirma que transcurrido mucho tiempo la temperatura se iguala a la temperatura ambiente.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos
tiempo = np.array([0, 2, 4, 6, 8, 10, 12])
temperatura = np.array([43.7, 37.7, 35.7, 34.4, 33.2, 32.9, 32.2])
temp_ambiente = 22

# Función de enfriamiento de Newton
def enfriamiento_newton(t, k, T0):
    return temp_ambiente + (T0 - temp_ambiente) * np.exp(-k * t)

# Ajustar los parámetros
popt, _ = curve_fit(enfriamiento_newton, tiempo, temperatura)
k, T0 = popt

print(f"Constante de enfriamiento k: {k:.4f} min^-1")
print(f"Temperatura inicial T0 (ajustada): {T0:.2f}°C")
print(f"Temperatura inicial real: 43.7°C")

# Calcular temperatura a los 16 minutos
T_16 = enfriamiento_newton(16, k, 43.7)
print(f"Temperatura después de 16 minutos: {T_16:.2f}°C")

# Calcular temperatura a largo plazo
t_largo_plazo = 1000  # Un tiempo muy grande
T_largo_plazo = enfriamiento_newton(t_largo_plazo, k, 43.7)
print(f"Temperatura a largo plazo: {T_largo_plazo:.2f}°C")

# Generar puntos para la curva de ajuste
t_fit = np.linspace(0, 60, 200)
temp_fit = enfriamiento_newton(t_fit, k, 43.7)

# Graficar
plt.figure(figsize=(12, 6))
plt.plot(tiempo, temperatura, 'bo', label='Datos experimentales')
plt.plot(t_fit, temp_fit, 'r-', label='Modelo de enfriamiento')
plt.axhline(y=temp_ambiente, color='g', linestyle='--', label='Temperatura ambiente')
plt.title('Enfriamiento')
plt.xlabel('Tiempo (minutos)')
plt.ylabel('Temperatura (°C)')
plt.legend()
plt.grid(True)
plt.show()

# Función para la sucesión
def sucesion_temperatura(n):
    return temp_ambiente + (43.7 - temp_ambiente) * (1 - k)**n

# Calcular y mostrar los primeros términos de la sucesión
n_terms = 10
for n in range(n_terms):
    print(f"T_{n} = {sucesion_temperatura(n):.2f}°C")



# Calcular el tiempo para acercarse a 22.1°C (muy cerca de la temperatura ambiente)
tiempo_cercano = np.log((22.1 - temp_ambiente) / (43.7 - temp_ambiente)) / -k
print(f"\nTiempo para acercarse a 22.1°C: {tiempo_cercano:.2f} minutos")
```

En la gráfica que compartiste, se puede observar la evolución de la temperatura de un sistema a lo largo del tiempo, comparando los datos experimentales (puntos azules) con un modelo teórico de enfriamiento basado en la Ley de Enfriamiento de Newton (curva roja). En el eje x se representa el tiempo en minutos, mostrando cómo varía la temperatura del sistema en el eje y, medido en grados Celsius. Los puntos azules corresponden a los datos experimentales que reflejan la disminución de la temperatura con el tiempo, mientras que la línea roja es la curva del modelo teórico ajustado utilizando la constante de enfriamiento  $ k = 0.066 $, que indica un proceso de enfriamiento relativamente rápido.

Una constante de enfriamiento $ k = 0.066 $ indica un enfriamiento mucho más rápido que en los casos típicos de enfriamiento del agua, que suele estar en el rango de $0.01$ a $0.04$  un valor $ k = 0.066$ se podría observar en situaciones donde el calor se disipa de manera muy eficiente.

La línea verde punteada marca la temperatura ambiente $(~22°C)$, que es el valor al cual se aproxima el sistema conforme pasa el tiempo. Este comportamiento es típico en procesos de enfriamiento donde la temperatura del objeto tiende hacia la temperatura del entorno.



##2.5
###**Explicación sobre alcanzar 10°C o 5°C:**

En el escenario en que se realizó el experimento, la temperatura ambiente era de 22°C. Dado que las temperaturas de 10°C a 5°C están por debajo de este valor, es imposible que el agua las alcance según el modelo planteado. Esto se debe a que, según la Ley de Enfriamiento de Newton, la temperatura ambiente representa una asíntota, a la cual la temperatura del agua se acercará indefinidamente desde arriba, sin descender en ningún momento por debajo de ella.

Desde una perspectiva termodinámica, el hecho de que el agua no pueda alcanzar temperaturas inferiores a la temperatura ambiente se relaciona con el segundo principio de la termodinámica, que establece que el calor fluye espontáneamente desde un cuerpo caliente hacia uno frío, pero no en la dirección opuesta. En este caso, dado que el agua es más caliente que el ambiente, transfiere calor al entorno hasta que se alcanza un equilibrio térmico. Este equilibrio se diferencia del modelo matemático de Newton, en el cual la temperatura ambiente es tratada como una asíntota que nunca se alcanza por completo. En la realidad, el agua eventualmente llegará a la misma temperatura que el ambiente, momento en el cual el flujo de calor se detiene. Sin un mecanismo activo, como un refrigerador u otro medio de enfriamiento, no es posible extraer calor adicional del agua, por lo que no puede enfriarse por debajo de la temperatura ambiente, ya que ello violaría las leyes termodinámicas.
